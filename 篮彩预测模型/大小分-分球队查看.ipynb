```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter
%matplotlib inline
```

```python
df_group = df.groupby('大小分')
df_group.mean()[['AST','AST_Away','BLK','BLK_Away','PF','PF_Away','FG3A','FG3A_Away']]
#从结果可以看出 大小不同的比赛的各项数据差别不大
```

```python
a = df_group.get_group('大')['hostTeamName'].value_counts()
b = df_group.get_group('大')['awayTeamName'].value_counts()
c = df_group.get_group('小')['hostTeamName'].value_counts()
d = df_group.get_group('小')['awayTeamName'].value_counts()
```

```python
pd.DataFrame([a,b,c,d],index=['主场大','客场大','主场小','客场小']).T.sort_values(by='主场大')
```
