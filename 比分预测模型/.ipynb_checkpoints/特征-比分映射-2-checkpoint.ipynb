- 最后发现，只需要FGM、FG3M、FTM这3个特征（从第一次模型中的coef_中也可以看出，只有三个参数的权重起作用），就可以很好的完成预测比分任务<br\>

补充：FGM、FG3M、FTM和PTS存在已知的线性关系，其实用不到数据挖掘。即挖掘以后发现的是很明显的线性关系。

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel("../数据收集/games.xlsx")
df.head()
```

```python
df.columns
```

```python
train_num = round(df.shape[0]*0.8)
y = df['PTS'].get_values()
y_train = y[:train_num]
y_test = y[train_num:]

x = df[['FGM','FG3M','FTM']].get_values()
x_train = x[:train_num]
x_test = x[train_num:]

scaler = StandardScaler().fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
y_predict = reg.predict(x_test)
y_predict-y_test
```
