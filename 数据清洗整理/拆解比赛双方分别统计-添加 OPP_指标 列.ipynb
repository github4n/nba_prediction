```python
import pandas as pd
import numpy as np
import os
```

```python
records = [f for f in os.listdir() if 'records' in f]
columns = ['SEASON_ID', 'TEAM_ID', 'TEAM_ABBREVIATION', 'TEAM_NAME', 'GAME_ID',
       'GAME_DATE', 'MATCHUP', 'WL', 'MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M',
       'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST',
       'STL', 'BLK', 'TOV', 'PF', 'PTS', 'PLUS_MINUS','TEAM_NAME_中文']
columns_away = [col+'_Away' for col in columns]
columns2 = ['jcTotalScore','总分', '大小分', '比赛双方']
df_new = pd.DataFrame()
for f in records[-1:]:
    df = pd.read_excel(f)
    teams = df['hostTeamName'].unique()
    for team in teams[:1]:
        df_new = pd.concat(df_new,df[df['TEAM_NAME_中文']==team][columns+columns2].values)
        df_new = pd.concat(df_new,df[df['TEAM_NAME_中文_Away']==team][columns_away+columns2].values)
```
